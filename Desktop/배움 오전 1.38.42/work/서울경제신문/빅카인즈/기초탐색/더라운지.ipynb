## 1. ì´ìŠˆ íë¦„ & ì´ìŠˆ ë§µ

```python
import requests
import os
import time
from dotenv import load_dotenv
from collections import Counter, defaultdict

# âœ… STEP 0: í™˜ê²½ ì„¸íŒ…
load_dotenv()
API_KEY = os.getenv("BIGKINDS_KEY")

search_url = "https://tools.kinds.or.kr/search/news"
feature_url = "https://tools.kinds.or.kr/feature"

# âœ… STEP 1: ë‰´ìŠ¤ ìˆ˜ì§‘
query = "ê³µê¸‰ë§ AND ì‚°ì—… AND ìˆ˜ì¶œ"
from_date = "2025-04-01"
until_date = "2025-04-015"
fetch_count = 30

print("\nğŸ“¥ STEP 1: ë‰´ìŠ¤ ìˆ˜ì§‘ ì¤‘...")
search_payload = {
    "access_key": API_KEY,
    "argument": {
        "query": query,
        "published_at": {
            "from": from_date,
            "until": until_date
        },
        "return_size": fetch_count,
        "return_from": 0,
        "fields": ["title", "news_id", "published_at"]
    }
}
search_resp = requests.post(search_url, json=search_payload)
docs = search_resp.json()["return_object"]["documents"]
news_ids = [doc["news_id"] for doc in docs]
titles = [doc["title"] for doc in docs]
dates = [doc["published_at"][:10] for doc in docs]  # YYYY-MM-DD

# âœ… STEP 2: ë³¸ë¬¸ ì¡°íšŒ + í‚¤ì›Œë“œ ì¶”ì¶œ
print("\nğŸ” STEP 2: í‚¤ì›Œë“œ ì¶”ì¶œ ì¤‘...")
view_payload = {
    "access_key": API_KEY,
    "argument": {
        "news_ids": news_ids,
        "fields": ["title", "content"]
    }
}
view_resp = requests.post(search_url, json=view_payload)
articles = view_resp.json()["return_object"]["documents"]

full_articles = []
all_keywords = []

for i, article in enumerate(articles):
    title = article["title"]
    content = article["content"]
    date = dates[i]

    # íŠ¹ì„± ì¶”ì¶œ
    feature_payload = {
        "access_key": API_KEY,
        "argument": {
            "title": title,
            "sub_title": "",
            "content": content
        }
    }
    f_resp = requests.post(feature_url, json=feature_payload)
    keyword_list = []
    if f_resp.status_code == 200:
        try:
            raw = f_resp.json()["return_object"]["result"]["title"]
            keyword_list = [k.split("|")[0] for k in raw.split(" ") if "|" in k]
            all_keywords.extend(keyword_list)
        except:
            pass

    full_articles.append({
        "title": title,
        "date": date,
        "keywords": keyword_list,
        "content": content
    })

# âœ… STEP 3: ì „ì²´ í‚¤ì›Œë“œ ë¶„ì„ + ê¸°ì‚¬ ê·¸ë£¹í™”
print("\nğŸ“Š STEP 3: ì´ìŠˆ ê·¸ë£¹ í‚¤ì›Œë“œ ë¶„ì„")
counter = Counter(all_keywords)
top_keywords = [kw for kw, _ in counter.most_common(5)]
print("Top í‚¤ì›Œë“œ:", ", ".join(top_keywords))

# ê¸°ì‚¬ ê·¸ë£¹í•‘
grouped_articles = {k: [] for k in top_keywords}
for article in full_articles:
    for k in top_keywords:
        if k in article["keywords"]:
            grouped_articles[k].append(article)

# âœ… STEP 4: ë‚ ì§œ ê¸°ë°˜ íë¦„ ë¶„ì„
print("\nğŸ“ˆ STEP 4: ë‚ ì§œë³„ íë¦„ ìš”ì•½")
date_map = defaultdict(list)
for art in full_articles:
    date_map[art["date"]].append(art)

for date in sorted(date_map.keys()):
    daily = date_map[date]
    d_keywords = []
    for a in daily:
        d_keywords.extend(a["keywords"])
    top_d = Counter(d_keywords).most_common(3)
    print(f"\nğŸ“… {date} ({len(daily)}ê±´)")
    print("- ì£¼ìš” í‚¤ì›Œë“œ:", ", ".join([k for k, _ in top_d]))
    print("- ëŒ€í‘œ ê¸°ì‚¬:", daily[0]["title"][:60], "...")
    print("- ì „ì²´ ê¸°ì‚¬:")
    for art in daily:
        print("  â€¢", art["title"])

    print("\nğŸ“ ê¸°ì‚¬ ë³¸ë¬¸ ìš”ì•½:")
    for art in daily:
        print(f"[{art['title']}]")
        print(art['content'][:300], "...\n")
```

## 2. ì´ìŠˆ êµ°ì§‘

```python
import os
import requests
import time
from dotenv import load_dotenv
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np

# âœ… STEP 0: í™˜ê²½ ì„¤ì •
load_dotenv()
API_KEY = os.getenv("BIGKINDS_KEY")
search_url = "https://tools.kinds.or.kr/search/news"

# âœ… STEP 1: ë‰´ìŠ¤ ìˆ˜ì§‘
query = "ì œì¡°ì—… AND ì‚°ì—… AND ì •ì±…"
from_date = "2025-04-01"
until_date = "2025-04-14"
fetch_count = 30

print("\nğŸ“¥ ë‰´ìŠ¤ ìˆ˜ì§‘ ì¤‘...")
search_payload = {
    "access_key": API_KEY,
    "argument": {
        "query": query,
        "published_at": {
            "from": from_date,
            "until": until_date
        },
        "return_size": fetch_count,
        "return_from": 0,
        "fields": ["title", "news_id", "published_at"]
    }
}
search_resp = requests.post(search_url, json=search_payload)
docs = search_resp.json()["return_object"]["documents"]

news_ids = [doc["news_id"] for doc in docs]
titles = [doc["title"] for doc in docs]
dates = [doc["published_at"][:10] for doc in docs]

# âœ… STEP 2: ê¸°ì‚¬ ë³¸ë¬¸ í™•ë³´
print("\nğŸ“„ ë³¸ë¬¸ ìˆ˜ì§‘ ì¤‘...")
view_payload = {
    "access_key": API_KEY,
    "argument": {
        "news_ids": news_ids,
        "fields": ["title", "content"]
    }
}
view_resp = requests.post(search_url, json=view_payload)
articles = view_resp.json()["return_object"]["documents"]

texts = []
meta = []
for i, article in enumerate(articles):
    text = article["title"] + "\n" + article["content"]
    texts.append(text)
    meta.append({"title": article["title"], "date": dates[i]})

# âœ… STEP 3: TF-IDF ë²¡í„°í™”
print("\nğŸ” í…ìŠ¤íŠ¸ ë²¡í„°í™” ì¤‘...")
vectorizer = TfidfVectorizer(max_df=0.8, min_df=2, stop_words='english')
tfidf_matrix = vectorizer.fit_transform(texts)
terms = vectorizer.get_feature_names_out()

# âœ… STEP 4: KMeans êµ°ì§‘í™”
print("\nğŸ§  KMeans ê¸°ë°˜ êµ°ì§‘í™”...")
num_clusters = 4
km = KMeans(n_clusters=num_clusters, random_state=42)
km.fit(tfidf_matrix)
labels = km.labels_

# âœ… STEP 5: ê²°ê³¼ ì¶œë ¥
print("\nğŸ“¦ ë‰´ìŠ¤ êµ°ì§‘í™” ê²°ê³¼:")
groups = {}
for i, label in enumerate(labels):
    if label not in groups:
        groups[label] = []
    groups[label].append({"title": meta[i]["title"], "text": texts[i]})

for label, articles in groups.items():
    print(f"\nğŸ”¹ í´ëŸ¬ìŠ¤í„° {label} ({len(articles)}ê±´)")
    cluster_texts = [a["text"] for a in articles]
    cluster_matrix = vectorizer.transform(cluster_texts)
    summed = np.array(cluster_matrix.sum(axis=0)).flatten()
    top_indices = summed.argsort()[::-1][:5]
    top_keywords = [terms[i] for i in top_indices]
    print("- ì£¼ìš” í‚¤ì›Œë“œ:", ", ".join(top_keywords))
    for a in articles:
        print(f" - {a['title']}")
```

```python
import requests
import os
import time
from dotenv import load_dotenv
from collections import Counter, defaultdict

# âœ… STEP 0: í™˜ê²½ ì„¸íŒ…
load_dotenv()
API_KEY = os.getenv("BIGKINDS_KEY")

search_url = "https://tools.kinds.or.kr/search/news"
feature_url = "https://tools.kinds.or.kr/feature"

# âœ… STEP 1: ë‰´ìŠ¤ ìˆ˜ì§‘
query = "ë¬¸í™” AND ì—¬í–‰"
from_date = "2025-04-01"
until_date = "2025-04-14"
fetch_count = 30

print("\nğŸ“¥ STEP 1: ë‰´ìŠ¤ ìˆ˜ì§‘ ì¤‘...")
search_payload = {
    "access_key": API_KEY,
    "argument": {
        "query": query,
        "published_at": {
            "from": from_date,
            "until": until_date
        },
        "return_size": fetch_count,
        "return_from": 0,
        "fields": ["title", "news_id", "published_at"]
    }
}
search_resp = requests.post(search_url, json=search_payload)
docs = search_resp.json()["return_object"]["documents"]
news_ids = [doc["news_id"] for doc in docs]
titles = [doc["title"] for doc in docs]
dates = [doc["published_at"][:10] for doc in docs]  # YYYY-MM-DD

# âœ… STEP 2: ë³¸ë¬¸ ì¡°íšŒ + í‚¤ì›Œë“œ ì¶”ì¶œ
print("\nğŸ” STEP 2: í‚¤ì›Œë“œ ì¶”ì¶œ ì¤‘...")
view_payload = {
    "access_key": API_KEY,
    "argument": {
        "news_ids": news_ids,
        "fields": ["title", "content"]
    }
}
view_resp = requests.post(search_url, json=view_payload)
articles = view_resp.json()["return_object"]["documents"]

full_articles = []
all_keywords = []

for i, article in enumerate(articles):
    title = article["title"]
    content = article["content"]
    date = dates[i]

    # íŠ¹ì„± ì¶”ì¶œ
    feature_payload = {
        "access_key": API_KEY,
        "argument": {
            "title": title,
            "sub_title": "",
            "content": content
        }
    }
    f_resp = requests.post(feature_url, json=feature_payload)
    keyword_list = []
    if f_resp.status_code == 200:
        try:
            raw = f_resp.json()["return_object"]["result"]["title"]
            keyword_list = [k.split("|")[0] for k in raw.split(" ") if "|" in k]
            all_keywords.extend(keyword_list)
        except:
            pass

    full_articles.append({
        "title": title,
        "date": date,
        "keywords": keyword_list,
        "content": content
    })

# âœ… STEP 3: ì „ì²´ í‚¤ì›Œë“œ ë¶„ì„ + ê¸°ì‚¬ ê·¸ë£¹í™”
print("\nğŸ“Š STEP 3: ì´ìŠˆ ê·¸ë£¹ í‚¤ì›Œë“œ ë¶„ì„")
counter = Counter(all_keywords)
top_keywords = [kw for kw, _ in counter.most_common(5)]
print("Top í‚¤ì›Œë“œ:", ", ".join(top_keywords))

# ê¸°ì‚¬ ê·¸ë£¹í•‘
grouped_articles = {k: [] for k in top_keywords}
for article in full_articles:
    for k in top_keywords:
        if k in article["keywords"]:
            grouped_articles[k].append(article)

# âœ… STEP 4: ë‚ ì§œ ê¸°ë°˜ íë¦„ ë¶„ì„
print("\nğŸ“ˆ STEP 4: ë‚ ì§œë³„ íë¦„ ìš”ì•½")
date_map = defaultdict(list)
for art in full_articles:
    date_map[art["date"]].append(art)

for date in sorted(date_map.keys()):
    daily = date_map[date]
    d_keywords = []
    for a in daily:
        d_keywords.extend(a["keywords"])
    top_d = Counter(d_keywords).most_common(3)
    print(f"\nğŸ“… {date} ({len(daily)}ê±´)")
    print("- ì£¼ìš” í‚¤ì›Œë“œ:", ", ".join([k for k, _ in top_d]))
    print("- ëŒ€í‘œ ê¸°ì‚¬:", daily[0]["title"][:60], "...")
    print("- ì „ì²´ ê¸°ì‚¬:")
    for art in daily:
        print("  â€¢", art["title"])

    print("\nğŸ“ ê¸°ì‚¬ ë³¸ë¬¸ ìš”ì•½:")
    for art in daily:
        print(f"[{art['title']}]")
        print(art['content'][:300], "...\n")
```

## 2. ì´ìŠˆ êµ°ì§‘


```python
import os
import requests
import time
from dotenv import load_dotenv
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np

# âœ… STEP 0: í™˜ê²½ ì„¤ì •
load_dotenv()
API_KEY = os.getenv("BIGKINDS_KEY")
search_url = "https://tools.kinds.or.kr/search/news"

# âœ… STEP 1: ë‰´ìŠ¤ ìˆ˜ì§‘
query = "ë¬¸í™” AND ì—¬í–‰"
from_date = "2025-04-01"
until_date = "2025-04-14"
fetch_count = 30

print("\nğŸ“¥ ë‰´ìŠ¤ ìˆ˜ì§‘ ì¤‘...")
search_payload = {
    "access_key": API_KEY,
    "argument": {
        "query": query,
        "published_at": {
            "from": from_date,
            "until": until_date
        },
        "return_size": fetch_count,
        "return_from": 0,
        "fields": ["title", "news_id", "published_at"]
    }
}
search_resp = requests.post(search_url, json=search_payload)
docs = search_resp.json()["return_object"]["documents"]

news_ids = [doc["news_id"] for doc in docs]
titles = [doc["title"] for doc in docs]
dates = [doc["published_at"][:10] for doc in docs]

# âœ… STEP 2: ê¸°ì‚¬ ë³¸ë¬¸ í™•ë³´
print("\nğŸ“„ ë³¸ë¬¸ ìˆ˜ì§‘ ì¤‘...")
view_payload = {
    "access_key": API_KEY,
    "argument": {
        "news_ids": news_ids,
        "fields": ["title", "content"]
    }
}
view_resp = requests.post(search_url, json=view_payload)
articles = view_resp.json()["return_object"]["documents"]

texts = []
meta = []
for i, article in enumerate(articles):
    text = article["title"] + "\n" + article["content"]
    texts.append(text)
    meta.append({"title": article["title"], "date": dates[i]})

# âœ… STEP 3: TF-IDF ë²¡í„°í™”
print("\nğŸ” í…ìŠ¤íŠ¸ ë²¡í„°í™” ì¤‘...")
vectorizer = TfidfVectorizer(max_df=0.8, min_df=2, stop_words='english')
tfidf_matrix = vectorizer.fit_transform(texts)
terms = vectorizer.get_feature_names_out()

# âœ… STEP 4: KMeans êµ°ì§‘í™”
print("\nğŸ§  KMeans ê¸°ë°˜ êµ°ì§‘í™”...")
num_clusters = 4
km = KMeans(n_clusters=num_clusters, random_state=42)
km.fit(tfidf_matrix)
labels = km.labels_

# âœ… STEP 5: ê²°ê³¼ ì¶œë ¥
print("\nğŸ“¦ ë‰´ìŠ¤ êµ°ì§‘í™” ê²°ê³¼:")
groups = {}
for i, label in enumerate(labels):
    if label not in groups:
        groups[label] = []
    groups[label].append({"title": meta[i]["title"], "text": texts[i]})

for label, articles in groups.items():
    print(f"\nğŸ”¹ í´ëŸ¬ìŠ¤í„° {label} ({len(articles)}ê±´)")
    cluster_texts = [a["text"] for a in articles]
    cluster_matrix = vectorizer.transform(cluster_texts)
    summed = np.array(cluster_matrix.sum(axis=0)).flatten()
    top_indices = summed.argsort()[::-1][:5]
    top_keywords = [terms[i] for i in top_indices]
    print("- ì£¼ìš” í‚¤ì›Œë“œ:", ", ".join(top_keywords))
    for a in articles:
        print(f" - {a['title']}")
```
